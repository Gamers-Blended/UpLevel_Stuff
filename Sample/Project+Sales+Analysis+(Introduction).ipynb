### Step 4: Read your Excel file into a DataFrame
There are two ways to proceed from here:

### Step 5: Check for sum of nulls in each column

```python
# Step 5: Check for nulls in each column
# To get the number of missing values in each column:
df.isnull().sum()
```

### Step 6: Set index with InvoiceDate
You should see that only Description and Customer ID columns are missing. 

Next, you will replace the index with the InvoiceDate column.

<strong>Hint: "use column as index pandas"</strong>

```python
# Step 6: Set index as InvoiceDate
df.set_index('InvoiceDate', inplace=True)
```

### Step 7: Import plotting libraries
Now that you've an DateTimeIndex, i.e. index that contains DateTime, you can now perform time-series analysis.

Import the following libraries:
1. matplotlib.pyplot as plt
2. seaborn as sns

```python
# Step 7: Import plotting libraries
import seaborn as sns
import matplotlib.pyplot as plt
```

### Step 8: Plot time-series analysis for Quantity
Now that you've the DateTime as an index, you can plot a time-series plot of Quantity.

You can use either seaborn or matplotlib.pyplot!

```python
# Step 8: Plot time-series for Quantity
sns.distplot(df['Quantity'])
plt.show()
```

### Step 9: Remove negative values
Looks like there's a massive oddity here - there are negative Quantity values here.

To clean this, we will remove all rows with negative quantities in the DataFrame.

<strong>Hint: Google "remove rows with certain values pandas"</strong>

```python
# Step 9: Remove rows with negative Quantity
df = df[df['Quantity'] > 0]
```

### Step 10: Plot Quantity again
With the negative values removed, let's try plotting the graph again, i.e. repeat Step 8.

```python
# Step 10: Plot time-series for Quantity
sns.distplot(df['Quantity'])
plt.show()
```

### Step 11: Resample the DataFrame by day and obtain median
If we plot the graph as it is, it presents a few challenges:
1. the time stamps are repeated, i.e. multiple purchases are happening
2. the range in purchase quantity is large
3. the timescale is too granular

As such, we should take the median of either each hour/day's purchase to plot instead. 

Resampling is useful in this case, where we get the median/mean of the timescale involved. If you do resampling properly and obtain the median for each day, you would see this:



<strong>Hint: Google "resample dataframe by day median"</strong>

```python
# Step 11: Resample your data by day, and pick the mean/median
resampled_mean = df.resample('D').mean()
resampled_median = df.resample('D').median()

resampled_mean
```

### Step 12: Check for missing data
Again, we can't be too sure whether your DataFrame after resampling is perfect (doesn't hurt to check).

```python
# Step 12: Check for null rows in columns
resampled_mean.isnull().sum()
```

### Step 13: Remove missing rows in the resample DataFrame
Aren't you glad you checked? Remove the rows containing NA.

```python
# Step 13: Remove rows with missing data
resampled_mean = resampled_mean.dropna()
```

### Step 14: Plot the resample DataFrame
Now that you've cleared the empty rows, time to repeat the plotting.

```python
# Step 14: Plot the resample DataFrame
sns.distplot(resampled_mean['Quantity'])
plt.show()
```

Well done - you've performed rudimentary data cleaning and obtained a better perspective of the median/mean sales over time.

### Step 15: Look at what is being sold the most
Back to the original DataFrame, let's ask two questions:
1. What was the most popular item, i.e. bought by most number of unique customers?
2. What was the item that is bought in the largest quantity?

To answer both questions, what you can use is .groupby() method - it is a really powerful way to group your data together for operations, e.g., summing, counting, averaging, etc. 

To answer question 1, you will have to perform a groupby by the "Description" column, and count the number of unique values in each column. 



To answer question 2, you will have to perform a groupby by the "Description" column, and sum instead to get the total number of quantities purchased for each description/item. 



<strong>Hint: Google "groupby count pandas"</strong>

```python
df = pd.concat(pd.read_excel('online_retail_II.xlsx', sheet_name=None), ignore_index=True)
```

```python
df2 = df.copy()
# Step 15a: Groupby description and then sum the groupby DataFrame, and sort by Customer ID
df2 = df2.groupby(by='Description').count()
df2.sort_values(by='Customer ID', axis=0, ascending=False, inplace=True)
df2
```

```python
df3 = df.copy()
# Step 15b: Groupby description and then sum the groupby DataFrame, and sort by Quantity
df3 = df3.groupby(by='Description').sum()
df3.sort_values(by='Quantity', axis=0, ascending=False, inplace=True)
df3
```
