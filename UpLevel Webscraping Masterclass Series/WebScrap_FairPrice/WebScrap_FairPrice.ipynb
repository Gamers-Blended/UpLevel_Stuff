# Scale Out to Page 1 to Page 40

```python
# each & in the URL specifies a different query
# starting from the question mark in the URL
# So splitting the url apart makes it easier to work with since we can keep the base_url
base_url = "https://website-api.omni.fairprice.com.sg/api/product/v2?category=fruits-vegetables&experiments=recHome-Z%2CbsHome-Z%2CpastPurchaseCategory-B%2CsearchVariant-B%2CUnappliedPromoVariant-B%2CtrendH-Z%2CtimerVariant-Z%2CsubstitutionVariant-B%2CsubstitutionVariant-B%2CsubstitutionVariant-B%2CsubstitutionVariant-B&includeTagDetails=true&page="
end_url = "&pageType=category&slug=fruits-vegetables&storeId=165&url=fruits-vegetables"
```

```python
df_list = []

# loop 
# page 0 same as page 1, hence start at pg1
for page_num in range(1,40):
    # construct temporary URL
    temp_url = base_url + str(page_num) + end_url
    
    # make a GET request
    temp_res = requests.get(temp_url)
    
    # get the JSON from response object
    temp_json = temp_res.json()
    
    # get DataFrame
    temp_df = pd.json_normalize(temp_json['data']['product'])
    
    # append the df into the list of dfs
    df_list.append(temp_df)
    
# combine df in the list of dfs, into a huge df
final_df = pd.concat(df_list)
# reset index
final_df.reset_index(drop=True, inplace=True)
final_df
```

```python
store_data = final_df['storeSpecificData']

store_list = []

for store in store_data:
    store_temp = pd.json_normalize(store[0])
    store_list.append(store_temp)
    
final_store_df = pd.concat(store_list)
final_store_df.reset_index(drop=True, inplace=True)
final_store_df
```

```python
final_combined_df = pd.concat([final_df, final_store_df], axis=1)
final_combined_df
```

```python
final_combined_df.to_csv("Dataset/fair_price_final_combined.csv", index=None)
```

# Analysis

```python
# keep only these xolumns
col_list = ["name",
           "inStoreStock",
           "mrp",
           "metaData.Country of Origin",
           "brand.name"]

subset_df = final_combined_df[col_list]
subset_df
```

```python
ax = sns.countplot(x = subset_df['metaData.Country of Origin'])
ax.set(xlabel='Country of Origin', ylabel='Count')
plt.xticks(rotation=90)
sns.despine()
```

```python
plt.figure(figsize = (16,9))
plt.barh(y = subset_df['metaData.Country of Origin'].value_counts().sort_values().index,
    width = subset_df['metaData.Country of Origin'].value_counts().sort_values())
plt.title("Country of Origin")
```

```python
# to show effects of sns.despine()
plt.figure(figsize = (16,9))
plt.barh(y = subset_df['metaData.Country of Origin'].value_counts().sort_values().index,
    width = subset_df['metaData.Country of Origin'].value_counts().sort_values())
sns.despine() # remove top and right spines from plot
plt.title("Country of Origin")

# to export png file
plt.savefig("CountryOfOrigin.png")
```
