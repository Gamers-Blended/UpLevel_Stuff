```python
categories = []
titles = []
urls = []
timestamp = []

counter = 0

# loop through all 15 items
for item in test_div:
# for item in test_ol:
#     temp_category = item.find("a", class_ = "teaser__category").text
    temp_category = item["data-category"]
    categories.append(temp_category)
    
#     temp_title = item.find("a", class_ = "teaser__title").text
    temp_title = item["data-heading"]
    titles.append(temp_title)
    
#     temp_url = "https://www.channelnewsasia.com" + item.find("a", class_ = "teaser__title")['href']
    temp_url = item["data-link_absolute"]
    urls.append(temp_url)
    
#     temp_timestamp = int(item.find("time")['datetime'])
    temp_timestamp = int(test_time[counter]["data-lastupdated"])
    timestamp.append(temp_timestamp)
    
    counter += 1
```

```python
temp_df = pd.DataFrame({"category": categories,
                       "title": titles,
                       "url": urls,
                       "timestamp": timestamp})

temp_df
```

```python
test_first_url = temp_df['url'][0]
test_first_url
```

```python
temp_article = []
articles = []

for url in temp_df['url']:
    temp_article_res = requests.get(url, headers=headers)
#     temp_article_res = requests.get(url)
    temp_article_soup = bs4.BeautifulSoup(temp_article_res.text)
#     temp_article_div = temp_article_soup.find("div", class_ = "c-rte--article")
    temp_article_div = temp_article_soup.find("div", class_ = "text-long")
    temp_article_p = temp_article_div.find_all("p")
    
    temp_clean_article_p = []
    
    for p in temp_article_p:
        temp_clean_article_p.append(p.text.strip().replace("\xa0", " "))
        
    temp_article = " ".join(temp_clean_article_p)
    
    articles.append(temp_article)
```

```python
temp_df['article'] = articles
```

```python
temp_article_res = requests.get(temp_df['url'][0], headers=headers)
# temp_article_res = requests.get(temp_df['url'][0])

temp_article_res
```
