# Simple Test Using Remote Webdriver

```python
options = webdriver.ChromeOptions()
options.add_argument('--headless')
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')

driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)
driver.get("https://www.google.com")
```

```python
# initialise webdriver object
# local webdriver
# driver = webdriver.Chrome(PATH_TO_WEBDRIVER)
# remote webdriver
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

# open the browser
driver.get("https://mobile.facebook.com/groups/DataScienceGroup")

# wait a bit for the page to load completely
time.sleep(5)

# click on the accept essential cookie button
accept_cookies_button = driver.find_element(By.CLASS_NAME, "_56bt")
accept_cookies_button.click()

# login
username_textbox = driver.find_element(By.CLASS_NAME, "_8qtn")
username_textbox.send_keys(FB_USERNAME)
password_textbox = driver.find_element(By.CLASS_NAME, "_8qtm")
password_textbox.send_keys(FB_PASSWORD)
login_button = driver.find_element(By.CLASS_NAME, "_56bu")
login_button.click()

# scroll to the bottom of site
driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

time.sleep(2)

# get the cross button
# cross_button = driver.find_element(By.CLASS_NAME, "_4b-n")
# click the whatever element that selenim found
# cross_button.click()

time.sleep(3)
driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

# get the "See More..." button
# see_more = driver.find_element(By.CLASS_NAME, "m_more_item")

# click see more
# see_more.click()

# get the page source
source = driver.page_source

# need to logout

# close the driver
driver.close()
```

```python
test_soup = bs4.BeautifulSoup(source)
```

```python
# find the storyStream section
story_container = test_soup.find("div", id = "m_group_stories_container")
story_container
```

```python
article_list = story_container.find_all("article", class_ = "_55wo _5rgr _5gh8 async_like")
article_list
```

# Simple Test With Xpath

```python
from selenium.webdriver.common.keys import Keys
import random

# initialise webdriver object
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
driver.get("https://www.google.com")

# wait a bit for the page to load completely
time.sleep(3)

# click on reject all cookies button
# reject_all_button = driver.find_element(By.CLASS_NAME, "tHlp8d")
reject_all_button = driver.find_element(By.XPATH, "//button[@class='tHlp8d']")
reject_all_button.click()

# find search box
input_box = driver.find_element(By.XPATH, "//input[@class='gLFyf gsfi']")

# add characters into search box
for char in "ABC":
    input_box.send_keys(char)
    # add some delay between key inputs
    time.sleep(random.randint(1,2))

# start search
input_box.send_keys(Keys.ENTER)
```
